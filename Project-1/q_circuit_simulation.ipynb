```python
import numpy as np

# Import Qiskit
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram, plot_state_city
import qiskit.quantum_info as qi
```

```python
def simulate_circuit(circuit):
    if not isinstance(circuit, QuantumCircuit):
        raise TypeError("Expected a QuantumCircuit object from qiskit")
    else:
        simulator = AerSimulator()
        circuit = transpile(circuit, simulator)
        result = simulator.run(circuit).result()
        counts = result.get_counts(circuit)
        return counts
```

```python
# Test the simulator with a simple circuit that generates a Bell pair
bell_state = QuantumCircuit(2)
bell_state.h(0)
bell_state.cx(0,1)
bell_state.measure_all()

simulate_circuit(bell_state)

hist_data = simulate_circuit(bell_state)
print(hist_data)
plot_histogram(hist_data, title="Measurment Histogram")
```
